# Exctract

```python
def extract_characters_attributes(movie_plot):
    doc = nlp(nlp(movie_plot)._.resolved_text)

    characters_attributes = {}

    for ent in doc.ents:
        if ent.label_ in ["PERSON"]:
            character_name = ent.text.strip()
            attributes = set()
            agent_verbs = set()
            patient_verbs = set()

            for token in ent.subtree:
                if token.dep_ in ["amod"] or token.pos_ in ["ADJ", "NOUN"]:
                    attributes.add(token.lemma_)
                if token.dep_ in ["nsubj"] and token.head.pos_ == "VERB":
                    agent_verbs.add(token.head.lemma_)
                if (token.dep_ in ["dobj", "nsubjpass", "iobj"] or (token.dep_ == "prep" and token.head.dep_ == "pobj")) and token.head.pos_ == "VERB":
                    patient_verbs.add(token.head.lemma_)

            for child in ent.root.head.subtree:
                if child.dep_ in ["amod"] or child.pos_ in ["ADJ"]:
                    attributes.add(child.lemma_)
                if child.dep_ in ['nsubj'] and child.pos_ == "VERB":
                    agent_verbs.add(child.head.lemma_)
                if (child.dep_ in ["dobj", "nsubjpass", "iobj"] or (child.dep_ == "prep" and child.head.dep_ == "pobj")) and child.pos_ == "VERB":
                    patient_verbs.add(child.head.lemma_)

            exists=0

            for existing_character in characters_attributes.keys():
                if character_name in existing_character or existing_character in character_name:
                    exists=1
                    attributes.update(characters_attributes[existing_character]["attributes"])
                    agent_verbs.update(characters_attributes[existing_character]["agent_verbs"])
                    patient_verbs.update(characters_attributes[existing_character]["patient_verbs"])
                    characters_attributes[existing_character] = {"attributes": list(attributes), "agent_verbs": list(agent_verbs), "patient_verbs": list(patient_verbs)}

            if exists == 0:
                characters_attributes[character_name] = {"attributes": list(attributes), "agent_verbs": list(agent_verbs), "patient_verbs": list(patient_verbs)}

    return characters_attributes
```

```python
import torch
import json
import os

resolved = []

df = pd.read_csv('/content/drive/MyDrive/latentPersonas/movie_plots_complete.csv')


total_plots = len(df)


output_dir = '/content/drive/MyDrive/latentPersonas/extracted_info'
os.makedirs(output_dir, exist_ok=True)


if 'filename' not in df.columns:
    df['filename'] = ""


for index, row in tqdm(df.iterrows(), total=total_plots, desc="Processing plots"):


    if not pd.isnull(row['filename']) and row['filename'] != "":
        continue  # Skip this row if the 'filename' is not empty

    torch.cuda.empty_cache()

    title = row['title']
    letter = row['letter']
    plot = row['plot']

    file_number = str(index + 1).zfill(5)
    filename = f"{letter}_{file_number}.json"



    file_path = os.path.join(output_dir, filename)
    with open(file_path, 'w') as fp:
        try:
            if len(plot.split()) <= 1:
                json.dump({}, fp)
            else:
                json.dump(extract_characters_attributes(plot), fp)
        except:
            json.dump({}, fp)


    df.at[index, 'filename'] = filename
    df.to_csv('/content/drive/MyDrive/latentPersonas/movie_plots_complete.csv', index=False)
```
